```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.describe()
```

```python
glucose_mean = df['Glucose'].mean()
BloodPlessure_mean = df['BloodPressure'].mean()
insulin_mean = df['Insulin'].mean()
BMI_mean = df['BMI'].mean()
```

```python
df['Glucose'] = df['Glucose'].replace(0, glucose_mean)
df['BloodPressure'] = df['BloodPressure'].replace(0, BloodPlessure_mean)
df['Insulin'] = df['Insulin'].replace(0, insulin_mean)
df['BMI'] = df['BMI'].replace(0, BMI_mean)
```

```python
df.describe()
```

```python
df['Insulin'] = np.where(df['Insulin'] > 400, insulin_mean, df['Insulin'])
```

```python
X = df[['Glucose', 'BloodPressure', 'Insulin', 'BMI', 'Age', 'DiabetesPedigreeFunction']]
y = df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, test_size=0.3, random_state=0)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_train)
```

```python
lr = LogisticRegression(C=1000, random_state=0)
lr.fit(X_train_std, y_train)
lr.predict_proba(X_test_std)
```

```python
y_pred = lr.predict(X_test_std)
```

```python
i = 0
count = 0
for x, x1 in zip(y_pred, y_test):
    if x == x1:
        count += 1
        
# accuracy_score(y_test, y_pred)
```

```python
print(count / len(y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_pred, y_test)
```
